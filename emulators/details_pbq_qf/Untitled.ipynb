```python
import numpy  as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame.from_csv('photobleaching_mixture00_grid.csv')
```

```python
print(data)
```

```python
degs = np.asarray(data['0.189174654'])
```

```python
sns.distplot(degs)
```

```python
sorted_degs = np.sort(degs)
```

```python
fig = plt.figure(figsize = (12, 2))
sns.rugplot(sorted_degs, height = 0.5)
plt.xlim(-0.05, 0.1)
plt.show()
```

```python
fig = plt.figure(figsize = (12, 2))
sns.kdeplot(sorted_degs, bw = 0.0005, gridsize = 5000)
plt.xlim(-0.05, 0.1)
plt.ylim(-0.01, 8.0)
plt.show()
```
